```python
condd =1
conds = 4
ls = 3e+4
ld = 1e+3
Vs = 1e+9
Vd  = 1e+10
ks = 1e+9
kd = 1e+9
pd0 = 0e+7
pd0_lin = 0e+7
pt = 2e+7
mu = 100


taus = 1e+7
taud = 7e+6
rho = 2600
g = 9.8
Rcyl = 5e+2
S = 3.14 * Rcyl**2 
R1 = rho * g * Vs /(ks*S)
T1 = (conds / condd )**4 * ld /ls
un_plus_T1 = 1 + T1
un_plus_R1 = 1 + R1

phi = kd /ks * Vs / Vd
tstar = Vs * 8 * mu * ld / (ks * 3.14 * condd**4)
ps_analytic = []
pd_analytic = []
ps_linearized =[]
pd_linearized = []
time =[]
time_linearized = []
n = 1
numerator = taus + R1 * n * (taus - taud) / un_plus_R1 + pd0 / un_plus_T1 -pt * T1 / un_plus_T1
denominator = taus + R1 * (n-1) * (taus - taud) / un_plus_R1 - 2 * (taus - taud) / un_plus_R1 + pd0 / un_plus_T1 -pt * T1 / un_plus_T1
tend = - tstar / un_plus_T1 * np.log(numerator/denominator)
tend_lin = tend
t0 = 0
t0_lin = 0
while not np.isnan(tend):
#while n < 3:
    print(n)
    t = np.linspace(0,tend,100)
    t_lin = np.linspace(0,tend_lin,100)
    ps0 = - taus - (n -1) * R1 * (taus - taud) / un_plus_R1 + 2 * (taus - taud) / (1 + R1)
    ps = pd0 * 1 / un_plus_T1 - pt * T1 / un_plus_T1 + (ps0 - pd0 * 1 / un_plus_T1 + pt * T1 / un_plus_T1) * np.exp( - un_plus_T1 * t/tstar )
    pd = pd0 + kd * 3.14 * condd**4 / (Vd * 8 * mu * ld)*(-pd0 + pd0 / un_plus_T1 - pt * T1 / un_plus_T1) * t  
    + phi / un_plus_T1 * (ps0 - pd0 / un_plus_T1 + pt * T1 / un_plus_T1) * (np.exp(- un_plus_T1 * t /tstar) - 1)
    ps_lin = pd0_lin * 1 / un_plus_T1 - pt * T1 / un_plus_T1 + (ps0 - pd0_lin * 1 / un_plus_T1 + pt * T1 / un_plus_T1) * np.exp( - un_plus_T1 * t/tstar )
    pd_lin = pd0_lin - phi/tstar *( pd0_lin - ps0) * t                                                                                                                                               
    t = t + t0
    t_lin = t_lin + t0_lin
    pd0 = pd[-1]
    pd0_lin = pd_lin[-1]
    t0 = t[-1] 
    t0_lin = t_lin[-1]
    numerator = taus + R1 * n * (taus - taud) / un_plus_R1 + pd0 / un_plus_T1 -pt * T1 / un_plus_T1
    denominator = taus + R1 * (n-1) * (taus - taud) / un_plus_R1 - 2 * (taus - taud) / un_plus_R1 + pd0 / un_plus_T1 -pt * T1 / un_plus_T1
    numerator_lin = taus + R1 * n * (taus - taud) / un_plus_R1 + pd0_lin / un_plus_T1 -pt * T1 / un_plus_T1
    denominator_lin = taus + R1 * (n-1) * (taus - taud) / un_plus_R1 - 2 * (taus - taud) / un_plus_R1 + pd0_lin / un_plus_T1 -pt * T1 / un_plus_T1
    tend = - tstar / un_plus_T1 * np.log(numerator/denominator)
    tend_lin = - tstar / un_plus_T1 * np.log(numerator_lin/denominator_lin)
    ps_analytic.append(ps)
    pd_analytic.append(pd)
    ps_linearized.append(ps_lin)
    pd_linearized.append(pd_lin)
    time.append(t)
    time_linearized.append(t_lin)
    n = n + 1
time = np.concatenate(time,axis = 0)
time_linearized = np.concatenate(time_linearized,axis = 0)

ps_analytic = np.concatenate(ps_analytic,axis = 0)
pd_analytic = np.concatenate(pd_analytic,axis = 0)
ps_linearized = np.concatenate(ps_linearized,axis = 0)
pd_linearized = np.concatenate(pd_linearized,axis = 0)
len(pd),len(ps_analytic)
```

```python
poisson = 0.25
lame = 1e+9
x = np.linspace(-10000,10000,200)
y = np.linspace(-8000,8000,150)
X,Y = np.meshgrid(x,y)
xSource = np.array([0,3000])
ySource =  np.array([0,-3000])
depthSource = np.array([1000,3000])
V = np.array([Vs,Vd])
deltaP = np.concatenate((ps,pd),axis = 1)
radial_distance = np.zeros((2,np.shape(X)[0],np.shape(X)[1]))
radial_distance[0,:,:] = np.sqrt((X - xSource[0])**2 + (Y - ySource[0])**2)
radial_distance[1,:,:] = np.sqrt((X - xSource[1])**2 + (Y - ySource[1])**2)
uzMax = np.array([3./(3.14* 4) * V[0] * (1 - poisson) / lame * deltaP[:,0] / depthSource[0]**2, 3./(3.14* 4) * V[1] * (1 - poisson) / lame * deltaP[:,1]/ depthSource[1]**2] )
uz = np.zeros((len(time),2,np.shape(radial_distance)[1],np.shape(radial_distance)[2]))

for i in np.arange(len(time)):
    uz[i,0,:,:] = uzMax[0,i] * ( 1 + (radial_distance[0,:,:] / depthSource[0])**2)**(- 3./2)
    uz[i,1,:,:] = uzMax[1,i] * ( 1 + (radial_distance[1,:,:] / depthSource[1])**2)**(- 3./2)
uz[:,0,:,:] = uz[:,0,:,:] - uz[0,0,:,:]
uz[:,1,:,:] = uz[:,1,:,:] - uz[0,1,:,:]
uztot = uz[:,0,:,:] + uz[:,1,:,:]
```
