```python
import numpy as np
import matplotlib.pyplot as plt
poisson = 0.25
lame = 1e+9
elastic_constant = (1 - poisson) / lame
x = np.array([0])
y = np.array([0])
xi = np.array([1,0.01])
xSource = np.array([0,0])
ySource = np.array([0,0])
VSource = np.array([1e+9,5e9])
depthSource = np.array([1000,5000])
t = np.linspace(0,20)
deltaP = np.zeros((len(t),len(xSource)))
deltaP[:,0] = 1e+6 * np.exp(-xi[0] * t)
deltaP[:,1] = 1e+6 * np.exp(-xi[1] * t)
uz = np.zeros((len(x),len(t),len(xi)))
uz_total = np.zeros((len(t),len(x)))
for countStation in np.arange(len(x)):
    for countSource in np.arange(np.shape(deltaP)[1]):
        R = np.sqrt((x[countStation] - xSource[countSource])**2 + (y[countStation] - ySource[countSource])**2 + depthSource[countSource] **2)
        coeff_up = depthSource[countSource] / R **3 * elastic_constant
        uz[countStation,:,countSource] = deltaP[:,countSource] * VSource[countSource] * coeff_up
    uz_total[:,countStation] = np.sum(uz[countStation,:,:],axis = 1)
plt.plot(t,uz_total[:,0]- uz_total[0,0] ,'o')
plt.plot(t,uz[0,:,0] - uz[0,0,0],'b')
plt.plot(t,uz[0,:,1] - uz[0,0,1],'r')
```
